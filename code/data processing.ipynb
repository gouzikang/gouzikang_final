```python
import pandas as pd
df = pd.read_csv('1.csv')

df['shou'] = (df['close'] - df['open'])/df['open']
df_2014_2015 = df[(df['date']>= '2014-01-01') & (df['date']< '2016-01-01')]
df_2016 = df[(df['date']>= '2016-01-01') & (df['date']< '2017-01-01')]

# make sure rows of 2 tables contain same 'symbol' value, otherwise filter them out
new_df_2014_2015 = pd.pivot_table(df_2014_2015.dropna(), values='shou', index=['symbol'], columns=['date'])
new_df_2016 = pd.pivot_table(df_2016.dropna(), values='shou', index=['symbol'], columns=['date'])

# filter out rows with symbols not present in both tables
symbols_2014_2015 = set(new_df_2014_2015.index)
symbols_2016 = set(new_df_2016.index)
common_symbols = symbols_2014_2015.intersection(symbols_2016)
new_df_2014_2015 = new_df_2014_2015[new_df_2014_2015.index.isin(common_symbols)]
new_df_2016 = new_df_2016[new_df_2016.index.isin(common_symbols)]


# filter out rows with symbols not present in both tables and delete rows in two final tables where there is nan in rows
new_df_2014_2015 = new_df_2014_2015.dropna()
new_df_2016 = new_df_2016.dropna()

# delete rows in two final tables where there is nan in rows
# filter out rows with symbols not present in both tables
symbols_2014_2015 = set(new_df_2014_2015.index)
symbols_2016 = set(new_df_2016.index)
common_symbols = symbols_2014_2015.intersection(symbols_2016)
new_df_2014_2015 = new_df_2014_2015[new_df_2014_2015.index.isin(common_symbols)]
new_df_2016 = new_df_2016[new_df_2016.index.isin(common_symbols)]

# Transpose the two tables
new_df_2014_2015 = new_df_2014_2015.transpose()
new_df_2016 = new_df_2016.transpose()
```
